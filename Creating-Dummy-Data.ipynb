```python
#library
import numpy as np
import matplotlib.pyplot as plt
```

# Problem 1
Creating random numbers

```python
np.random.seed(0)
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
my_rand = np.random.multivariate_normal(mean, cov, 500)
print(my_rand.shape)
```

# Problem 2
Scatter plot

```python
plt.scatter(my_rand[:,0],my_rand[:,1])
plt.title("Scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

# Problem 3
Histogram

```python
fg, axes = plt.subplots(2, 1, sharex=True,figsize=(8,10))
axes[0].hist(my_rand[:,0], bins=70)
axes[0].set_title('Histogram of x1')
plt.setp(axes[0], xlabel='x1')
plt.setp(axes[0], ylabel='frequency')
axes[0].set_ylim([0,28])
axes[1].hist(my_rand[:,1], bins=70)
axes[1].set_title('Histogram of x2')
plt.setp(axes[1], xlabel='x2')
plt.setp(axes[1], ylabel='frequency')
axes[1].set_ylim([0,28])
plt.show()
```

# Problem 4
More data

```python
np.random.seed(0)
mean2 = (0,-3)
cov = [[1.0, 0.8], [0.8, 1.0]]
my_rand2 = np.random.multivariate_normal(mean2, cov, 500)
print(my_rand2.shape)
```

```python
plt.scatter(my_rand[ : , 0],my_rand[:,1],color="blue",label="0")
plt.scatter(my_rand2[ : , 0],my_rand2[:,1],color="orange",label="1")
plt.title("Scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.show()
```

# Problem 5
Data combination

```python
combined = np.concatenate((my_rand,my_rand2),axis=0)
combined.shape
```

# Problem 6
Labeling

```python
# creating label column with 500 pcs of 0's and 500 pcs of 1's
label = np.concatenate((np.zeros((500,1)),np.ones((500,1))),axis=0)
dataset = np.concatenate((combined,label),axis=1)
print("Size of the dataset:",dataset.shape)
dataset
```
